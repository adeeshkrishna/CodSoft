Importing necessary libraries and packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,ConfusionMatrixDisplay,auc,roc_curve,accuracy_score
import warnings
```

```python
warnings.filterwarnings("ignore")
```

Loading the dataset

```python
df=pd.read_csv('CreditCardFraud.csv')
df
```

Information about the dataset

```python
df.info()
```

Describing the dataset

```python
df.describe()
```

Checking for missing values

```python
df.isna().sum()
```

Checking for duplicate values

```python
duplicate = df.loc[df.duplicated(keep = False)]
duplicate
```

Removing duplicate values

```python
df.drop_duplicates(inplace=True)
```

Checking the data type of each column

```python
df.dtypes
```

Checking the distribution of class labels

```python
sns.countplot(x=df["Class"],data=df,hue=df["Class"])
plt.title('Distribution of output class in data')
plt.xlabel('Fraud (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
df["Class"].value_counts()
```

Splitting input and output

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1]
```

Oversampling the dataset as it is highly imbalanced

```python
os=SMOTE(random_state=1)
X_os,y_os=os.fit_resample(X,y)
```

```python
print(X_os.shape)
print(y_os.shape)
```

Scaling the input using MinMax scaler

```python
scaler=MinMaxScaler()
X_os_scaled=scaler.fit_transform(X_os)
```

Splitting the dataset into training and testing dataset

```python
X_os_train,X_os_test,y_os_train,y_os_test=train_test_split(X_os_scaled,y_os,test_size=0.3,random_state=1)
```

Building LogisticRegression and RandomForestClassifier models

```python
acc=[]
rf=RandomForestClassifier(random_state=1)
lgr=LogisticRegression()
models=[rf,lgr]
for model in models:
  model.fit(X_os_train,y_os_train)
  y_pred=model.predict(X_os_test)
  acs=accuracy_score(y_os_test,y_pred)
  acc.append(acs*100)
  print(model)
  print(classification_report(y_os_test,y_pred))
  print('*'*55)
```

Comparing accuracy of RandomForestClassifier and LogisticRegression

```python
m=['RandomForestClassifier()','LogisticRegression()']
acd=pd.DataFrame({'Model':m,'Accuracy':acc})
acd
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x='Model',y='Accuracy',data=acd,hue='Model',legend=False)
plt.xticks(rotation=90)
plt.show()
```

Since RandomForestClassifier has the highest accuracy of 99.98 %, using that model for prediction purpose

```python
rf.fit(X_os_train,y_os_train)
```

Plotting confusion matrix

```python
y_os_pred=rf.predict(X_os_test)
print(classification_report(y_os_test,y_os_pred))
print(ConfusionMatrixDisplay.from_predictions(y_os_test,y_os_pred))
```

Plotting AUC-ROC curve

```python
y_pred_prob=rf.predict_proba(X_os_test)[:,-1]
```

```python
fpr,tpr,thresholds = roc_curve(y_os_test, y_pred_prob)
roc_auc = auc(fpr, tpr)
```

```python
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area ={roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
